# Basic Data Exploration

```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('Salaries.csv')
df.head()
```

```python
df.tail()
```

```python
df.columns
```

```python
df.shape #(number of rows , number of columns)
```

# Data Cleaning

```python
df.info()
```

```python
# drop Notes , Status
column_to_drop = ['Notes','Status']
df = df.drop(column_to_drop, axis=1)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Columns to fill with mean
columns_to_fill = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']

# Calculate the mean for each column
mean_values = df[columns_to_fill].mean()

# Fill missing values in each column with the corresponding mean
df[columns_to_fill] = df[columns_to_fill].fillna(mean_values)
# we use the mean becouse of -->  Ease of Implementation
```

```python
df.isnull().sum()
```

# Basic Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Histogram for the distribution of salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['BasePay'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('BasePay')
plt.ylabel('Frequency')
plt.show()
```

```python
department_counts = df['Agency'].value_counts()
department_counts
```

```python

# Pie chart for the proportion of employees in different departments
plt.figure(figsize=(8, 8))
department_counts = df['Agency'].value_counts()
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

# Grouped Analysis:

```python
department_counts = df['Year'].value_counts()
department_counts
```

```python
grouped_data = df.groupby(['Year'])

# Calculate summary statistics for each grou -->  mean salary
summary_statistics = grouped_data['BasePay'].mean()

# Display the summary statistics
print(summary_statistics)
```

# Simple Correlation Analysis:

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()
# Heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
correlation = df['BasePay'].corr(df['OvertimePay'])

print(f"Correlation between BasePay and OvertimePay: {correlation}")

# Scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BasePay', y='OvertimePay', data=df, alpha=0.7)
plt.title('Scatter Plot: BasePay vs OvertimePay')
plt.xlabel('BasePay')
plt.ylabel('OvertimePay')
plt.show()
```
